#### 1. Importando Librerias

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn import decomposition
```

```python
X1,Y1 =  make_blobs(n_features=10,n_samples=100,centers=4,random_state=4,cluster_std=1)
```

```python
print(X1.shape)
print(Y1.shape)
```

#### 2. Creacion de Dataframe

```python
df = pd.DataFrame(X1,index=range(X1.shape[0]),columns=range(X1.shape[1]))
```

LMPLOT dispersion o scatter

```python
pca = decomposition.PCA(n_components=4)
pc  = pca.fit_transform(X1)
```

```python
df =pd.DataFrame({'var': pca.explained_variance_ratio_,'PC':['PC1','PC2','PC3','PC4']})
```

```python
df
```

```python
plt.figure()
sns.barplot(x='PC',y='var',data=df,color='c')
```

```python
pc_df = pd.DataFrame(data=pc,columns=['PC1','PC2','PC3','PC4'])
pc_df['Cluster']=Y1
```

```python
sns.lmplot(x='PC1',y='PC2',
           data=pc_df,
           fit_reg=False,
           hue='Cluster',#Color by cluster
           legend=True,
           scatter_kws={'s':80})
plt.show()
```
